```python
import random
import networkx as nx
import matplotlib.pyplot as plt



# Генерация начальных позиций для роботов
initial_positions = [(random.uniform(0, 10), random.uniform(0, 10)) for _ in range(10)]
print("Начальные позиции:")
for i, pos in enumerate(initial_positions):
    print(f"Робот {i+1}: {pos}")
while True:
  try:
    object = tuple(float(x) for x in input("Введите позицию искомого объекта (10): ").split())
    if len(object) < 2:
      raise ValueError
    break
  except:
    print("Ошибка! Введите два числа через пробел, пожалуйста.\n")
rng = 2
initial_positions += [object]

# Создание графа для моделирования связей между роботами
G = nx.Graph()
G.add_nodes_from(range(len(initial_positions)))

# Добавление случайных связей между роботами
for i in range(len(initial_positions) - 1):
    for j in range(i+1, len(initial_positions)- 1):
        if random.random() < 0.5:  # Вероятность связи
            G.add_edge(i, j)

def calculate_fitness(position):
    if (position[0]-object[0])**2 + (position[1]-object[1])**2 > rng**2:
        return 0
    elif position[0] == object[0] and position[1]==object[1]:
        return 1
    else:
        return 1-((position[0]-object[0])**2 + (position[1]-object[1])**2) / rng**2

# Основной цикл работы
generation = 0
while True:
    print(f"\nПоколение {generation+1}:")

    # Вычисление "фитнеса" для каждой позиции
    fitness_scores = [calculate_fitness(pos) for pos in initial_positions[:-1]]
    if 1 not in fitness_scores:

        # Вывод "фитнеса" для каждой позиции
        for i, score in enumerate(fitness_scores):
            print(f"Фитнес робота {i}: {score}")

        # Обновление позиций для следующего поколения
        for i in range(len(initial_positions) - 1):
            if fitness_scores[i] == 0:
                initial_positions[i] = (abs(initial_positions[i][0] + random.uniform(-1, 1)), abs(initial_positions[i][1] + random.uniform(-1, 1)))
            else:
                initial_positions[i] = ((initial_positions[i][0] + object[0])/2 if (initial_positions[i][0] - object[0] > 0.1) else object[0], (initial_positions[i][1] + object[1])/2 if (initial_positions[i][1] - object[1] > 0.1) else object[1])

        # Вывод промежуточных позиций
        print("Промежуточные позиции:")
        for i, pos in enumerate(initial_positions[:-1]):
            print(f"Робот {i}: {pos}")

        # Обмен информацией между соседними роботами
        for node in G.nodes:
            neighbors = list(G.neighbors(node))
            if 11 in neighbors:
                neighbors.remove(11)
            if neighbors:
                best_neighbor = max(neighbors, key=lambda x: fitness_scores[x])
                if fitness_scores[best_neighbor] > fitness_scores[node]:
                    initial_positions[node] = ((initial_positions[node][0] - initial_positions[best_neighbor][0])/2 if (initial_positions[i][0] - initial_positions[best_neighbor][0] > 0.1) else initial_positions[best_neighbor][0], (initial_positions[i][1] + initial_positions[best_neighbor][1])/2 if (initial_positions[i][1] - initial_positions[best_neighbor][1] > 0.1) else initial_positions[best_neighbor][1])

        x, y = 10, 10
        x0, y0 = 0, 0
        fig, ax, ax0 = plt.subplot(), plt.subplot(), plt.subplot()
        ax.scatter(x, y)
        ax0.scatter(x0, y0)
        nx.draw(G, initial_positions, with_labels=True, node_color='lightblue', node_size=1500, font_weight='bold', font_size=12, edge_color='gray')

        plt.show()
        generation += 1
        if object in initial_positions[:-1]:
            print(f"Робот {initial_positions.index(object)} нашел искомый объект")
            break

        #zaglushka = input("Нажмите enter для продолжения")
    else:
        print(f"Робот {fitness_scores.index(1)} нашел искомый объект")
        break
```
